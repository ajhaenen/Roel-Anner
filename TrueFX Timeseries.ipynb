# Collect data from TrueFX
From a single month (January 2016), all 15 exchanges are downloaded from https://truefx.com/?page=download&description=january2016&dir=2016/JANUARY-2016
All files are unzipped and placed in folder Y:\Roel-Anner_DATA\TrueFX_201601\



Use TrueFX_csvread.m to take all seperate csv files and save a .mat file containing the Exchange structure.



Use theoretisch_maximum.m to determine the return of an ideal trading algorithm. For January 2016 on USDCHF with 1 pip extra spread this was 1470%.
